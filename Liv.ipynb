```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('LIV/RT_IOT2022.csv')
df.set_index('no', inplace = True)
df
```

```python
df.describe()
```

```python
dfref = df.iloc[1:-90000]
dfref.describe()
```

```python
dfcopy = dfref.groupby(['Attack_type'])['Attack_type'].count()
dfcopy
```

```python
# Pie chart of the number of attack types
plt.pie(dfcopy, autopct='%1.1f%%')
plt.show()
```
