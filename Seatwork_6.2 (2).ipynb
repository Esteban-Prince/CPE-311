```python
import pandas as pd

newdf = pd.read_csv('2019_Yellow_Taxi_Trip_Data.csv')
```

```python
mask = newdf.columns.str.contains('id$|store_and_fwd_flag', regex = True)
columns_to_drop = newdf.columns[mask]
columns_to_drop
```

```python
taxis = newdf.drop(columns=columns_to_drop)
taxis.head()
```

```python
taxis = taxis.rename(
    columns={
        'tpep_pickup_datetime': 'pickup_time',
        'tpep_dropoff_datetime': 'dropoff_time'
    }
)
taxis.columns
```

```python
taxis[['pickup_time','dropoff_time']] = taxis[['pickup_time','dropoff_time']].apply(pd.to_datetime)
taxis.dtypes
```

```python
taxis = taxis.assign(
    elapsed_time=lambda x: x.dropoff_time - x.pickup_time,
    cost_before_tip=lambda x: x.total_amount - x.tip_amount,
    tip_pct=lambda x: x.tip_amount / x.cost_before_tip,
    fees=lambda x: x.cost_before_tip - x.fare_amount,
    avg_speed=lambda x: x.trip_distance.div(x.elapsed_time.dt.total_seconds()/60/60
                                           )
)
```

```python
taxis.dtypes
```

```python
taxis.sort_values(['passenger_count','pickup_time'],ascending=[False,True]).head()
```

```python
taxis.head()
```

```python
taxis.nlargest(3,'elapsed_time')
```

```python
taxis.nlargest(3, 'trip_distance')
```
